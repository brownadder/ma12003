This notebook creates a zip file of your coursework for submission to Moodle.

## Steps 1 & 2 — Enter your student ID & Create Zip File

1. In the next cell, replace `"ps2106"` with your university ID.

```python
STUDENTID = "ps2106"
```

2. Then run the following cell to create your zip file.

```python
import os

workspace = os.path.expanduser("~/ma12003_workspace")
coursework_dir = os.path.join(workspace, "coursework")
zip_file = os.path.join(workspace, f"cw_{STUDENTID}.zip")

if not os.path.isdir(coursework_dir):
    raise FileNotFoundError(
        f"Could not find {coursework_dir}. "
        "Please make sure your coursework folder exists and contains your work."
    )

if os.path.exists(zip_file):
    os.remove(zip_file)

# KEY FIX: cd into workspace before zipping
!cd "{workspace}" && zip -r "cw_{STUDENTID}.zip" "coursework"

print(f"Created: {zip_file}")
```

## Step 3 — Locate your zip file & verify your submission

Your zip file has been created in your `ma12003_workspace` folder with the name:

`cw_STUDENTID.zip`

(where STUDENTID is your actual ID)

Run the next cell to check that:
- your files are included
- your working notebook works fully (or to the extent you have attempted the coursework)

Only proceed to submit if everything looks correct.

```python
import os
import shutil

workspace = os.path.expanduser("~/ma12003_workspace")
zip_file = os.path.join(workspace, f"cw_{STUDENTID}.zip")
extract_dir = os.path.join(workspace, "cw_VERIFY")

# 1. Check zip exists
if not os.path.isfile(zip_file):
    raise FileNotFoundError(f"Zip file not found: {zip_file}")

print(f"Found zip file: {zip_file}\n")

# 2. List contents
print("Contents of zip file:\n")
!unzip -l "{zip_file}"

# 3. Extract to clean folder
if os.path.exists(extract_dir):
    shutil.rmtree(extract_dir)

os.makedirs(extract_dir)

!unzip "{zip_file}" -d "{extract_dir}"

print(f"\nExtracted to: {extract_dir}")
```

## Step 4 — Submit to Moodle

1. Go to `ma12003_workspace` folder
2. Download your zip file (`cw_STUDENTID.zip`)
3. Upload it to Moodle

Make sure you upload the correct file. You may ignore the yellow triangle warning Moodle gives you for zip files.
