# **Mathematical Background: Inner Product Spaces**



A (real-valued) **inner product space** is a pair $(V, \langle \cdot, \cdot \rangle)$, where $V$ is a vector space over $\mathbb{R}$ equipped with a binary operation $\langle \cdot, \cdot \rangle : V \times V \to \mathbb{R}$ called the **inner product**, which satisfies the following axioms for all $u, v, w \in V$ and scalars $\lambda \in \mathbb{R}$:


$$\langle \lambda u + v, w \rangle = \lambda \langle u, w \rangle + \langle v, w \rangle$$     
$$\langle u, v \rangle = \langle v, u \rangle   $$                    
$$\langle v, v \rangle \geq 0 \quad \text{and} \quad \langle v, v \rangle = 0 \iff v = 0$$  



* Elements $u, v \in V$ of the inner product space are typically referred to as **vectors** (even when they do not have a geometric representation in 2D or 3D space).
* The norm induced by the inner product is defined as $\|v\| = \sqrt{\langle v, v \rangle}$ and represents the **length** or **magnitude** of a vector.
* The **unit vector** in the direction of a non-zero vector $v$ has length 1 and points in the same direction as $v$ and is defined as:

  $$
  \hat{v} := \frac{v}{\|v\|}
  $$

* The **projection** of a vector $v$ on a non-zero vector $w$ is defined as:

  $$
  \text{proj}_w(v) := \langle v, \hat{w} \rangle \cdot \hat{w}
  $$

  It measures the component of $v$ in the direction of $w$.

* The inner product $\langle u, v \rangle$ allows us to define the **angle** $\theta$ between two non-zero vectors $u$ and $v$ via the formula:

  $$
  u \wedge v := \cos \theta = \frac{\langle u, v \rangle}{\|u\| \cdot \|v\|}
  $$

  This generalises the familiar Euclidean definition of angle, and helps interpret the inner product as measuring how closely aligned two vectors are. 
  
  Note that the use of the notation $u \wedge v$ is non-standard, but will be used throughout this coursework. 



* If $\langle u, v \rangle = 0$, then $u$ and $v$ are said to be **orthogonal** (i.e., perpendicular).

#### **Example 1: Euclidean Space**

Let $V = \mathbb{R}^2$ be the set of all ordered pairs of real numbers. Define the inner product as:

$$
\langle (x_1, y_1), (x_2, y_2) \rangle = x_1 x_2 + y_1 y_2, \qquad (x_1, y_1), (x_2, y_2) \in \mathbb{R}^2.
$$

This is the standard dot product on the plane. It satisfies all inner product axioms and induces the usual Euclidean norm:

$$
\| (x, y) \| = \sqrt{x^2 + y^2}.
$$

**Remarks:** This space corresponds to the familiar 2D geometry. The inner product gives a natural interpretation of angle and projection.

#### **Example 2: Portfolio Space**

Let $V = \mathbb{R}^n$ represent the space of investment portfolios over $n$ assets. A portfolio $x \in V$ is a vector $(x_1, x_2, \dots, x_n)$ where $x_i \in \mathbb{R}$ denotes the holding (positive or negative) in asset $i$. Let $w = (w_1, w_2, \dots, w_n) \in \mathbb{R}^n$ be a fixed vector of **positive weights**, representing relative importance or sensitivity of each asset. Define the inner product:

$$
\langle x, y \rangle = \sum_{i=1}^n w_i x_i y_i.
$$

This is a **weighted inner product**, and it satisfies all axioms provided each $w_i > 0$.

**Remarks:** This inner product captures the degree of alignment between two portfolios while taking into account asset relevance via weights.

### **Task 2.c. Partition**

The **partition** of a list of points $Y=[v_1, \ldots, v_n]$ relative to a given list of $K$ centroids $C=[c_1, \ldots, c_K]$ assigns each point $v_i$ to the closest centroid $c_k$ using cosine dissimilarity, i.e. such that 

$$ \frac{1 - c_k \wedge v_i}{2} \leq \frac{1 - c_j \wedge v_i}{2}, \qquad j=1,\ldots,K $$

If a point is equally close to two or more centroids, it may be assigned to any one of them but not to both.

## **Task 3. Clustering for Inner Product Spaces** $~~~~$ [5 marks = 25%]

In this task, you will implement a cosine-based clustering algorithm for inner product spaces.

Your goal is to partition a given set of points (e.g. portfolios, R2 vectors, etc.) into $K \geq 2$ clusters, such that:

* each point belongs to exactly one cluster,
* each cluster has a centroid (also a vector of the same type),
* each cluster aims to minimises the cosine dissimilarity between its centroid and its members via the clustering algorithm described below. 

Implement the following method in the appropriate class(es):

```python
def clusters(initial_centroids, points, max_iter=100)
```

* **Inputs**:
  * `initial_centroids`: a list of length $K\geq 2$ (this list defines $K$) with points (i.e. vectors) from an inner product space,
  * `points`: a list of points from the same inner product space to be partitioned in clusters,
  * `max_iter`: an optional integer parameter specifying the maximum number of iterations of the clustering algorithms. 

* **Outputs**:
  * `clusters`: A list of length $K$, where each element is itself a list of vectors. These are the clusters your algorithm should have found. 
  * `centroids`: A list of length $K$ of centroids of these clusters. 


* **Clustering Algorithm**

    * **Partition Step**: Assign each point to the cluster whose centroid it is most aligned with, i.e. having least cosine dissimilarity.
    
    * **Update Step**: Compute the new centroid of each cluster.
    
    * **Convergence**: Repeat the above steps until a maximum number of iterations is reached.


**Hint:** Use `objectclass=type(centroids[0])` to dynamically find the class of centroids. Thereafter you may call `objectclass.partition` to call the `partition` method in `objectclass`, for example.
