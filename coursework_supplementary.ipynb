## Task 1. Caesar Cipher $~~~~$ [8 marks = 40%]

In this task you need to implement a class (i.e. user defined type) `CaesarCipher` that allows you to perform encoding and decoding using the Caesar Cipher described below.

### Introduction to the Caesar Cipher:
The Caesar Cipher is a cryptographic technique in which each letter in the text is shifted a certain number of places down or up the alphabet. The number of positions by which the letters are shifted is called the `key`.

#### Encryption Process
Choose a Key: Select a fixed number of positions to shift. For example, if the `key` is `3`, each letter will be shifted three places forward in the alphabet.

Shift each letter: For each letter in the text:

- Find its position in the alphabet (e.g., `A=0`, `B=1`, ..., `Z=25`).
- Add the `key` to this position.
- Find the letter that corresponds to the new position (modulo `26`, since there are `26` letters in the English alphabet).

For example, let's encrypt the text `"HELLO"` with a `key` of `3`.

- `H` -> `K` (`H` is the `8`th letter, `8+3=11`, `K` is the `11`th letter)
- `E` -> `H` (`E` is the `5`th letter, `5+3=8`, `H` is the `8`th letter)
- `L` -> `O` (`L` is the `12`th letter, `12+3=15`, `O` is the `15`th letter)
- `L` -> `O` (same as above)
- `O` -> `R` (`O` is the `15`th letter, `15+3=18`, `R` is the `18`th letter)

So, the text `"HELLO"` is encoded as `"KHOOR"`. 

- You should respect capitalisation, so that `"Hello"` is encoded as `"Khoor"`, i.e. the capital letter `H` is mapped to the capital letter `K`, while small letter `e` is mapped to the small letter `h`.
- You should leave non-alphabetical letters unchanged, so that `"Hello !!"` is encoded as `"Khoor !!"`. 
- You should only modify letters in the **English** alphabet, i.e. `A-Z` and `a-z`. *Unicode* (`utf-8` encoding, for example) and other formats of text encodings allow the use of non-English letters such as `ä`, `ó`, `ü`, `ç` etc. Such letters should not be modified. 

#### Decryption Process

The decryption process works by reversing the above process - by shifting each letter backward by the `key`, i.e. 

For each letter in the encrypted text:

- Find its position in the alphabet.
- Subtract the `key`.
- Find the letter that corresponds to the new position (modulo `26`).


After decryption, you should obtain the original text provided the same `key` was used for encryption and decryption.

We will illustrate the required encodings (i.e. rules for determining the `CaesarCipher` to be used) with the example of the following set of dialogues (which are stored in the file `'task2example.txt'`):

Each dialogue can consist of multiple lines of text spoken by one speaker. 

* Each dialogue in the play should be encoded using a `key` which is generated based on the name of the speaker speaking the dialogue. Specifically, you should consider all the letters appearing in the speaker's name and take the alphabetical position of the letter appearing the highest in the alphabet as the value of the `key`. We illustrate this with the following examples:

    $\diamond$ Example: The first dialogue is `'Hi Alice, do you still have the coin that I gave you?'`. This is spoken by `BOB`. The letter appearing the highest in the alphabest is `O`, which appears at position `14` in the alphabet, so this dialogue should be encoded with `key=14`.

    $\diamond$ Example: The last dialogue consists of two lines: `'Possibly.'` and `'But if I did, I would not return it to you.'`. This is spoken by `EVE`. The letter appearing the highest in the alphabest is `V`, which appears at position `21` in the alphabet, so these dialogues should be encoded with `key=21`.

* For determining the `key`, you should ignore all letters in the speaker's name that are not in the English alphabet.

* The lines with the names of the speakers (e.g. `'BOB.'`, `'ALICE.'` etc) should not be encoded or decoded. 

* Lines which are not dialogues, nor contain the name of a speaker, will be called *non-dialogues* in this coursework. These should be encoded with `key=-5` 

    $\diamond$ Example: We consider the lines `'This is an example to demonstrate Task 2.'`, `'Alice notices that Eve has just entered the room'`, and `'An uncomfortable silence ensues.'`, and blank or whitespace lines, to be *non-dialogues* and these should be encoded with `key=-5`.

#### Determining speakers and dialogues:


**Determining speakers**

* The line preceeding each speaker's name should have **no alphabetical characters**. It can have blank spaces, special character, numbers etc. 

* Each speaker's name appears **separately** on a **single line**, fully in **upper case**, and **ends with `'.'`**, with the possible exception of leading or trailing whitespace characters such as `' '`, `'\n'` etc.

* The speaker's name **can have non-alphabetical characters** as well as alphabetical characters that are not in the English alphabet. 

* Each speaker's name must have **at least 3 alphabetical characters** and **at most 20 alphabetical characters**.

**Determining dialogues**

* A character's dialogues starts in the **next line following the speaker's name**, but might span **multiple lines**.

* A character's dialogues is **ended by a line which has no alphabetical characters**. Such a line can have blank spaces, special character, numbers etc. This line is **NOT** considered a part of the dialogue.

* Note that a **multi-line dialogue** by a character is considered a **single dialogue**.
